```python
def get_correlation(data, threshold):
    corr_col = set()
    corrmat = data.corr()
    for i in range(len(corrmat.columns)):
        for j in range(i):
            if abs(corrmat.iloc[i, j])> threshold:
                colname = corrmat.columns[i]
                corr_col.add(colname)
    return corr_col
```

```python
# Objective function: Accuracy of a classifier based on selected features
average = "binary"


def fitness_function_best_params_SVM(selected_features, X_train, X_test, y_train, y_test):
    # Train a classifier using the selected features
        # 🔍 4. En iyi hiperparametreleri bulmak için GridSearchCV kullanımı
    param_grid = {
    'C': [0.1, 1, 10, 100],      # Düzenleme parametresi
    'gamma': ['scale', 'auto'],  # Gamma parametresi
    'kernel': ['linear', 'rbf']  # Kullanılacak çekirdek fonksiyonu
    }

    svm = SVC()
    grid_search = GridSearchCV(svm, param_grid, cv=5, scoring='accuracy')
    grid_search.fit(X_train, y_train)

    # 🎯 5. En iyi parametreleri alalım
    best_params_SVM = grid_search.best_params_
   
    
    # Return accuracy as fitness
    return best_params_SVM  

    
def fitness_function_k_best(selected_features, X_train, X_test, y_train, y_test):
    # Train a classifier using the selected features
    param_grid = {'n_neighbors': np.arange(1, 21)}  # 1'den 20'ye kadar K değerlerini test edelim
    knn = KNeighborsClassifier()
    grid_search = GridSearchCV(knn, param_grid, cv=5, scoring='accuracy')
    grid_search.fit(X_train[X_train.columns[selected_features]], y_train)

    # 🎯 5. En iyi K değerini alalım
    best_k = grid_search.best_params_['n_neighbors']
   
    
    # Return accuracy as fitness
    return best_k

def fitness_function_best_params(selected_features, X_train, X_test, y_train, y_test):
    
    param_grid = {
    'max_depth': [3, 5, 10, None],          # Maksimum derinliği sınırlandır
    'min_samples_split': [2, 5, 10],        # Dallanma için gereken minimum örnek sayısı
    'min_samples_leaf': [1, 2, 5],          # Yaprak düğümde bulunması gereken minimum örnek sayısı
    'max_features': ['sqrt', 'log2', None]  # Her bölünmede kullanılacak özellik sayısı
    }

    tree = DecisionTreeClassifier(random_state=42)
    grid_search = GridSearchCV(tree, param_grid, cv=5, scoring='accuracy')
    grid_search.fit(X_train[X_train.columns[selected_features]], y_train)

    # 🎯 4. En iyi parametreleri alalım
    best_params = grid_search.best_params_
    return best_params

    
#KNeighborsClassifier KNN
def fitness_function(selected_features, X_train, X_test, y_train, y_test):
    # Train a classifier using the selected features
    clf = KNeighborsClassifier(n_neighbors=best_k)
    clf.fit(X_train[:, selected_features], y_train)
    #clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test[:, selected_features])
    #y_pred = clf.predict(X_test)

    # Return accuracy as fitness
    return accuracy_score(y_test, y_pred)
    # Return accuracy as fitness
    #accuracy_score(y_test, y_pred)
    #precision_score(y_test, y_pred, average=average, zero_division=1)
    #f1_score(y_test, y_pred, average='binary', zero_division=1)
    #recall_score(y_test, y_pred, average=average, zero_division=1)




'''
#KNeighborsClassifier KNN
def fitness_function(selected_features, X_train, X_test, y_train, y_test):
    # Train a classifier using the selected features
    clf = KNeighborsClassifier(n_neighbors=best_k)
    clf.fit(X_train[:, selected_features], y_train)
    y_pred = clf.predict(X_test[:, selected_features])
    
    # Return accuracy as fitness
    return accuracy_score(y_test, y_pred)

#CART
def fitness_function(selected_features, X_train, X_test, y_train, y_test):
    # Train a classifier using the selected features
    clf =  DecisionTreeClassifier()
    clf.fit(X_train[:, selected_features], y_train)
    y_pred = clf.predict(X_test[:, selected_features])
    
    # Return accuracy as fitness
    return accuracy_score(y_test, y_pred)

    
#RandomForestClassifier
def fitness_function(selected_features, X_train, X_test, y_train, y_test):
    # Train a classifier using the selected features
    clf =RandomForestClassifier(
    n_estimators=42,         # 100 karar ağacı
    max_depth=5,              # Maksimum derinlik (aşırı büyümeyi önler)
    min_samples_split=4,      # Dallanma için minimum 4 örnek gerekir
    min_samples_leaf=2,       # Yaprakta en az 2 örnek bulunmalı
    max_features='sqrt',      # Her ağacın yalnızca sqrt(n) kadar özelliği kullanmasını sağlar
    random_state=42)
    clf.fit(X_train[:, selected_features], y_train)
    y_pred = clf.predict(X_test[:, selected_features])
    
    # Return accuracy as fitness
    return accuracy_score(y_test, y_pred)
    
#CART
def fitness_function(selected_features, X_train, X_test, y_train, y_test):
    # Train a classifier using the selected features
    clf =  DecisionTreeClassifier(**best_params, random_state=42)
    clf.fit(X_train[:, selected_features], y_train)
    y_pred = clf.predict(X_test[:, selected_features])
    
    # Return accuracy as fitness
    return accuracy_score(y_test, y_pred)


#SVM
def fitness_function(selected_features, X_train, X_test, y_train, y_test):
    # Train a classifier using the selected features
    clf = SVC(kernel='sigmoid', C=1.0, random_state=42)  # You can change the kernel
    clf.fit(X_train[:, selected_features], y_train)
    y_pred = clf.predict(X_test[:, selected_features])
    
    # Return accuracy as fitness
    #accuracy_score(y_test, y_pred)
    #precision_score(y_test, y_pred, average=average, zero_division=1)
    #f1_score(y_test, y_pred, average='binary', zero_division=1)
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average=average, zero_division=1)
    recall = recall_score(y_test, y_pred, average=average, zero_division=1)
    f1 = f1_score(y_test, y_pred, average=average, zero_division=1)
    
    return recall_score(y_test, y_pred, average=average, zero_division=1)


'''


# Initialize bee population
def initialize_population(num_bees, num_features):
    # Each bee is represented as a binary vector (1 = selected, 0 = not selected)
    return np.random.randint(2, size=(num_bees, num_features))

# Employed bee phase: Modify a solution and evaluate fitness
def employed_bee_phase(population, fitness, X_train, X_test, y_train, y_test):
    num_bees, num_features = population.shape
    new_population = population.copy()
    
    for i in range(num_bees):
        # Select a random feature to flip
        feature_to_flip = np.random.randint(num_features)
        new_solution = population[i].copy()
        new_solution[feature_to_flip] = 1 - new_solution[feature_to_flip]
        
        # Evaluate the new solution
        selected_features = np.where(new_solution == 1)[0]
        if len(selected_features) > 0:
            new_fitness = fitness_function(selected_features, X_train, X_test, y_train, y_test)
        else:
            new_fitness = 0
        
        # Update if the new solution is better
        if new_fitness > fitness[i]:
            new_population[i] = new_solution
            fitness[i] = new_fitness
    
    return new_population, fitness

# Onlooker bee phase: Probabilistic selection based on fitness
def onlooker_bee_phase(population, fitness, X_train, X_test, y_train, y_test):
    num_bees, num_features = population.shape
    new_population = population.copy()
    fitness_prob = fitness / fitness.sum()
    
    for _ in range(num_bees):
        # Select a bee probabilistically
        i = np.random.choice(num_bees, p=fitness_prob)
        
        # Select a random feature to flip
        feature_to_flip = np.random.randint(num_features)
        new_solution = population[i].copy()
        new_solution[feature_to_flip] = 1 - new_solution[feature_to_flip]
        
        # Evaluate the new solution
        selected_features = np.where(new_solution == 1)[0]
        if len(selected_features) > 0:
            new_fitness = fitness_function(selected_features, X_train, X_test, y_train, y_test)
        else:
            new_fitness = 0
        
        # Update if the new solution is better
        if new_fitness > fitness[i]:
            new_population[i] = new_solution
            fitness[i] = new_fitness
    
    return new_population, fitness

# Scout bee phase: Replace abandoned solutions
def scout_bee_phase(population, fitness, X_train, X_test, y_train, y_test, abandonment_limit):
    num_bees, num_features = population.shape
    for i in range(num_bees):
        if fitness[i] < abandonment_limit:
            # Generate a new random solution
            new_solution = np.random.randint(2, size=num_features)
            selected_features = np.where(new_solution == 1)[0]
            if len(selected_features) > 0:
                new_fitness = fitness_function(selected_features, X_train, X_test, y_train, y_test)
            else:
                new_fitness = 0
            
            # Update the population
            population[i] = new_solution
            fitness[i] = new_fitness
    
    return population, fitness

# ABC Algorithm
def artificial_bee_colony(X_train, X_test, y_train, y_test, num_bees=20, max_iter=10, abandonment_limit=0.1):
    num_features = X_train.shape[1]
    
    # Initialize population and evaluate fitness
    population = initialize_population(num_bees, num_features)
    fitness = np.zeros(num_bees)
    
    for i in range(num_bees):
        selected_features = np.where(population[i] == 1)[0]
        if len(selected_features) > 0:
            fitness[i] = fitness_function(selected_features, X_train, X_test, y_train, y_test)
    
    # Optimization loop
    for iteration in range(max_iter):
        # Employed bee phase
        population, fitness = employed_bee_phase(population, fitness, X_train, X_test, y_train, y_test)
        
        # Onlooker bee phase
        population, fitness = onlooker_bee_phase(population, fitness, X_train, X_test, y_train, y_test)
        
        # Scout bee phase
        population, fitness = scout_bee_phase(population, fitness, X_train, X_test, y_train, y_test, abandonment_limit)
        
        # Print the best solution in the current iteration
        best_solution = population[np.argmax(fitness)]
        best_fitness = np.max(fitness)
        print("----------------------------------------------------")
        print(f"Iteration {iteration + 1}: Best Fitness = {best_fitness}")
                
    # Return the best solution
    best_solution = population[np.argmax(fitness)]
    return best_solution
```
